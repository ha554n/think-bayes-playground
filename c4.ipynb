```python
euro = Euro(range(0, 101))
euro.plot()
```

```python
dataset = 'H' * 140 + 'T' * 110
euro.update_set(dataset)
euro.plot()
euro.summary()
```

```python
euro_tri = Euro()
for x in range(0, 51):
    euro_tri.set(x, x)
for x in range(51, 101):
    euro_tri.set(x, 100 - x)
euro_tri.normalize()
euro_tri.plot()
```
