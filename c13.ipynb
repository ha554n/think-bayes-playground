```python
import math
import random
from c13.util import make_cdf, fit_cdf, generate_cdf, cm_to_bucket
from c13.Calculator import Calculator
from c13.Calculator import plot_sequences
```

```python
interval = 3291.0
dt = 811.0 * 3
doublings = interval / dt
d1 = 15.5
d0 = d1 / 2.0 ** doublings
print("doublings = {:.2f}, d0 = {:.2f}".format(doublings, d0))
```

```python
d0 = 0.1
d1 = 15.5
doublings = math.log2(d1 / d0)
dt = interval / doublings
vdt = dt / 3
rdt = 365 / vdt
print("doublings = {:.2f}, dt = {:.2f}, vdt = {:.2f}, rdt = {:.2f}"
      .format(doublings, dt, vdt, rdt))
```

```python
cdf = make_cdf("original")
lam1 = fit_cdf(cdf)
fit = generate_cdf(lam1=lam1, name="fit")
cdf.plot_with([fit])
```

```python
calc = Calculator()
rho = 0.0
sequences = calc.make_sequences(100, rho, fit)
plot_sequences(sequences)
```

```python
cdf_2cm = calc.cache.conditional_cdf(cm_to_bucket(2), "2cm")
cdf_5cm = calc.cache.conditional_cdf(cm_to_bucket(5), "5cm")
cdf_10cm = calc.cache.conditional_cdf(cm_to_bucket(10), "10cm")
cdf_15cm = calc.cache.conditional_cdf(cm_to_bucket(15), "15cm")
cdf_2cm.plot_with([cdf_5cm, cdf_10cm, cdf_15cm])
```
